```python
# Step 11
import json
choice_description_items = {}
max_items = 0
max_item_name = ""
for index, row in chipo.iterrows():
    ingredients = row['choice_description']
    if type(ingredients) is not float:
        allIngredients = ingredients.split(',')
        for ing in allIngredients:
            actualIng = ing.replace("[", "")
            actualIng = actualIng.replace("]", "").strip()
            choice_description_items[actualIng] = choice_description_items.get(actualIng, 0) + 1
            if choice_description_items[actualIng] > max_items:
                max_items = choice_description_items[actualIng]
                max_item_name = actualIng

print(max_item_name, max_items)
```

```python
import matplotlib.pyplot as plt
```

```python
us_marriages_divorces = pd.read_csv('us-marriages-divorces-1867-2014.csv',sep=',')
```

```python
fig, ax = plt.subplots()
x_data = range(1867, 2012)
ax.plot(x_data, us_marriages_divorces['Marriages_per_1000'])
ax.plot(x_data, us_marriages_divorces['Divorces_per_1000'])
ax.set_title('Marriages and Divorces per capita in the US from 1867 to 2014')
ax.legend(['Marriages per 1000', 'Divorces per 1000'])
ax.set_xlabel('years')
ax.set_ylabel('Marriages/Divorces per 1000')
```

# Question 7

```python
x_values = np.array(['1900', '1950', '2000'])
marriages = []
divorces = []

for year in x_values:
    values = us_marriages_divorces.query(f'Year=={year}')
    marriages.append(values['Marriages_per_1000'].item())
    divorces.append(values['Divorces_per_1000'].item())

graph = pd.DataFrame({'Marriages per 1000': marriages,
                   'Divorces per 1000': divorces}, index=x_values)

ax = graph.plot.bar()
ax.set_xlabel("Years")
ax.set_ylabel("Marriages/Divorces per 1000")
ax.set_title('Marriages and Divorces per capita in the US in 1900, 1950 and 2000')
```
